```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Correct file path (as a string)
df = pd.read_csv(r"C:\Users\vivek\Downloads\archive.zip")

# Show first 5 rows
df.head()
```

```python
# Shape and column names
print("Shape:", df.shape)
print("Columns:", df.columns)

# Data types and non-null counts
df.info()

# Statistical summary (only for numerical columns)
df.describe()
```

```python
print("Missing values:\n", df.isnull().sum())
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
sns.countplot(x='gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
sns.histplot(df['math score'], kde=True, bins=20)
plt.title("Math Score Distribution")
plt.show()
```

```python
sns.countplot(x='lunch', data=df)
plt.title("Lunch Type Distribution")
plt.show()
```

```python
sns.boxplot(x='test preparation course', y='math score', data=df)
plt.title("Math Score vs Test Preparation Course")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='parental level of education', y='writing score', data=df)
plt.xticks(rotation=45)
plt.title("Writing Score vs Parental Education")
plt.show()
```

```python
sns.heatmap(df[['math score', 'reading score', 'writing score']].corr(), annot=True)
plt.title("Score Correlation Heatmap")
plt.show()
```
